```python
import tabula as tab
import pandas as pd
import gspread
import df2gspread as d2g
import pygsheets

gc = pygsheets.authorize(service_file='yourCreds.json')
sh = gc.open('yourSheet')
wb = sh[0]
```

```python
url = "https://www.wbhealth.gov.in/uploaded_files/corona/Vacant_bed_status_as_on_04.05_.2021_(1)_.pdf"
tables = tab.read_pdf(url, pages = "all", multiple_tables = True)

df_wb_test = pd.DataFrame()
```

```python
name = []

for item in tables[0]['Unnamed: 1'][1:21]:
    name.append(item)
    
for item in tables[1]['Name of CoVid Hospital'][0:22]:
    name.append(item)
    
for item in tables[2]['Name of CoVid Hospital'][0:22]:
    name.append(item)
    
for item in tables[3]['Name of CoVid Hospital'][0:22]:
    name.append(item)
    
for item in tables[4]['Name of CoVid Hospital'][0:10]:
    name.append(item)
    
vacant_num = []

for item in tables[0]['Unnamed: 3'][1:21]:
    vacant_num.append(item)
    
for item in tables[1]['Vacant\rBeds'][0:22]:
    vacant_num.append(item)    

for item in tables[2]['Vacant\rBeds'][0:22]:
    vacant_num.append(item)
    
for item in tables[3]['Vacant\rBeds'][0:22]:
    vacant_num.append(item)
    
for item in tables[4]['Vacant\rBeds'][0:10]:
    vacant_num.append(item)   

district = []

for item in tables[0]['Unnamed: 0'][1:21]:
    district.append(item)
    
for item in tables[1]['District'][0:22]:
    district.append(item)
    
for item in tables[2]['District'][0:22]:
    district.append(item)
    
for item in tables[3]['District'][0:22]:
    district.append(item)
    
for item in tables[4]['District'][0:10]:
    district.append(item)  
```

```python
df_wb_test['District'] = district
df_wb_test['Name'] = name
df_wb_test['Vacant'] = vacant_num
```
